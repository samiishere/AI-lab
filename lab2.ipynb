plotting histrogram for all columns

```python
data.hist(bins=15, figsize=(20, 15))
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()
```

display correlation heatmap

```python
plt.figure(figsize=(16, 12))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

pairplot for selected columns

```python
selected_columns = ['battery_power', 'ram', 'int_memory', 'px_height', 'px_width']
sns.pairplot(data[selected_columns], diag_kind='kde')
plt.suptitle('Pairplot of Selected Columns', y=1.02, fontsize=16)
plt.show()
```

Count plots for categorical columns

```python
categorical_columns = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']
plt.figure(figsize=(20, 12))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(2, 3, i)
    sns.countplot(data=data, x=column)
    plt.title(f'Count Plot of {column.capitalize()}')
plt.tight_layout()
plt.show()
```
